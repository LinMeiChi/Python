```python
# ----- 九九乘法表 -----
# 完整版
# 方法一: 利用 for 迴圈
print("------------------------- 方法一: -------------------------")
for i in range(1,10,1):
    for j in range(1,10,1):
        count = j * i
        print('%d*%d=%2d'%(j,i,count), end=' ')     # end 為結尾符號，不換行。
    print()
    
# 方法二: 利用 while 迴圈
print("\n------------------------- 方法二: -------------------------")
i=1    # 起始值
while i < 10:    # 條件
    j=1
    while j < 10:
        count =  i * j
        print('%d*%d=%2d'%(j,i,count), end=' ')
        j += 1
    i += 1    # 值更新
    print()
```

```python
# 左上三角格式輸出
for i in range(1,10):
    for j in range(1,10-i+1):
        print("%d*%d=%2d" % (j,i,i*j),end=" ")
    print("")
```

```python
# 左下三角格式輸出
for i in range(1,10):
    for j in range(1,i+1):
        print("%d*%d=%2d" % (j,i,i*j),end=" ")
    print("")
```

```python
# 右上三角格式輸出
# 方法一:
print("------------------------- 方法一: -------------------------")
for i in range(1,10):
    for j in range(1,i):
        print(end="       ")
    for j in range(i,10):
        print("%d*%d=%2d" % (j,i,i*j),end=" ")
    print("")
    
# 方法二:
print("\n------------------------- 方法二: -------------------------")
# 右上三角格式輸出
for i in range(1,10):
    for j in range(1,10):
        if j in range(1,i):
            print(end="       ")
        else:
            print("%d*%d=%2d" % (j,i,i*j),end=" ")
    print("")
```

```python
# 右下三角格式輸出
# 方法一:
print("------------------------- 方法一: -------------------------")
for i in range(1,10):
    for j in range(1,10-i):
        print(end="       ")
    for j in range(10-i,10):
        print("%d*%d=%2d" % (j,i,i*j),end=" ")
    print("")

# 方法二:
print("\n------------------------- 方法二: -------------------------")
for i in range(1,10):
    for j in range(1,10):
        if j in range(1,10-i):
            print(end="       ")
        else:
            print("%d*%d=%2d" % (j,i,i*j),end=" ")
    print("")
```
